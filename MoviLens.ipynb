```python
import numpy
from pandas import DataFrame, Series
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('Data_Sets/Movie_lens/users.dat', sep='::', header=None,
                        names=unames, engine = 'python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('Data_Sets/Movie_lens/ratings.dat', sep='::', header=None,
                        names=rnames, engine = 'python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('Data_Sets/Movie_lens/movies.dat', sep='::', header=None,
                         names=mnames, engine = 'python')
```

```python
users
```

```python
movies
```

```python
ratings
```

```python
df = pd.merge(pd.merge(movies, ratings), users)
```

```python
df
```

```python
mean = df.pivot_table(columns='gender', index='title', aggfunc='mean').rating
mean
```

```python
titles_by_count = df.groupby('title').rating.count()
titles_by_count
```

```python
active_titles = titles_by_count[titles_by_count >= 250]
```

```python
active_titles = active_titles.sort_values(ascending = False)
active_titles
```

```python
 mean_ratings = mean.loc[active_titles.index]
```

```python
mean_ratings.sort_values(by = 'F', ascending=False)
```

```python
mean_ratings.sort_values(by = 'M', ascending=False)
```

```python
mean_ratings['dissagreement'] = mean_ratings['M'] - mean_ratings['F']
```

```python
mean_ratings.sort_values('dissagreement', ascending = False)   # Movies like by men more than women
```

```python
mean_ratings.sort_values('dissagreement', ascending = True)   # Movies like by women more than men
```
